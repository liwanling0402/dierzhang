```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 
movies = ['', '', '4', '', '',
          '', '', '', '',
          '', '', '', '', '', '']
box_office = [50.35, 46.86, 42.50, 22.13, 17.16,
              15.58, 14.18, 13.30, 12.44, 11.38,
              10.25, 9.77, 9.29, 8.97, 8.16]

# 2. 
avg_box_office = np.mean(box_office)

# 3. 
bar_width = 0.5
y_pos = np.arange(len(movies))
plt.barh(y_pos, box_office, height=bar_width, label='')

# 4. 
plt.axvline(x=avg_box_office, linestyle='--', color='red', label=f'{avg_box_office:.2f}')

# 5. 
max_index = np.argmax(box_office)
plt.annotate('',
             xy=(box_office[max_index], y_pos[max_index]),
             xytext=(box_office[max_index]+5, y_pos[max_index]+0.2),
             arrowprops=dict(facecolor='black', shrink=0.05))

# 6. 
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = ["SimHei", "SimHei", "SimHei"]
plt.rcParams["axes.unicode_minus"] = False  # 
import matplotlib.pyplot as plt

# 6
ingredients = ['', '', '', '', '']
weights = [250, 100, 50, 30, 20]

# 
fig, ax = plt.subplots(figsize=(8, 6))
explode = (0.1, 0, 0, 0, 0)  # 
wedges, texts, autotexts = ax.pie(weights, explode=explode, labels=ingredients, autopct='%3.1f%%',
                                  startangle=90, textprops=dict(color="w"))

# 
ax.text(0, -1, '', fontsize=12, ha='center')

# 
table_data = [[ingredients[i], weights[i]] for i in range(len(ingredients))]
table = plt.table(cellText=table_data, colLabels=['', ''], loc='bottom', cellLoc='center', colLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width([0, 1])

# 
plt.subplots_adjust(bottom=0.2)
ax.axis('equal')  # 
plt.title('')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x[0, 2]
x = np.linspace(0, 2 * np.pi, 1000)
# sin(x)cos(x)
sin_x = np.sin(x)
cos_x = np.cos(x)

# sin(x)cos(x)
plt.plot(x, sin_x, label='sin(x)')
plt.plot(x, cos_x, label='cos(x)')

# axvspanx[0, /2]
plt.axvspan(0, np.pi/2, color='lightgreen', alpha=0.3)

# 1
plt.legend()
plt.title('')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], 
           ['0', '/2', '', '3/2', '2'])
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
