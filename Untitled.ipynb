```python
import matplotlib.pyplot as plt
# 解决中文显示问题
plt.rcParams["font.family"] = ["SimHei", "Arial"]  # 优先用SimHei，也可加Arial Unicode MS（mac系统常用）
plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示问题
import matplotlib.pyplot as plt
import numpy as np

# 还原图表数据（匹配Y轴600最大值、品类顺序）
categories = ['电子产品', '服装鞋帽', '家居百货', '美妆个护', '食品生鲜']
sales = [550, 420, 280, 390, 610]  # 销售额（万元），食品生鲜最高（610）符合图表

# 创建画布
plt.figure(figsize=(10, 6), dpi=100)

# 绘制柱形图（匹配图表蓝色柱形、无边框样式）
bars = plt.bar(
    x=categories,
    height=sales,
    color='#4472C4',  # 图表蓝色系
    width=0.6,
    edgecolor='none'
)

# 还原坐标轴与标签（匹配Y轴0-600范围、“好销”标注位置）
plt.ylim(0, 600)
plt.yticks(range(0, 601, 100))
plt.xlabel('商品品类', fontsize=12)
plt.ylabel('销售额（万元）', fontsize=12)
plt.title('电商平台5个品类月销售额统计', fontsize=14, fontweight='bold', pad=20)

# 还原“好销”文本标注（食品生鲜柱形旁）
plt.text(
    x=4, y=610 + 10, s='好销', 
    ha='center', va='bottom', 
    fontsize=11, fontweight='bold', color='red'
)

# 调整布局
plt.tight_layout()
plt.savefig('图表1_电商品类销售额柱形图.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 还原图表数据（匹配X轴0-14000范围、城市顺序自上而下）
cities = ['北京', '上海', '广州', '深圳']  # Y轴城市顺序（从上到下）
q1_spend = [12000, 11500, 9800, 10500]  # Q1支出
q2_spend = [13500, 12800, 10900, 11900]  # Q2支出（均高于Q1，符合图表）

# 设置条形位置（横向条形图Y轴索引）
y_pos = np.arange(len(cities))
bar_width = 0.35

# 创建画布
plt.figure(figsize=(10, 6), dpi=100)

# 绘制多组横向条形图（匹配图表Q1/Q2颜色、X轴范围）
plt.barh(
    y=y_pos - bar_width/2, width=q1_spend, 
    height=bar_width, label='Q1', color='#70AD47'  # 图表绿色系（Q1）
)
plt.barh(
    y=y_pos + bar_width/2, width=q2_spend, 
    height=bar_width, label='Q2', color='#ED7D31'  # 图表橙色系（Q2）
)

# 还原坐标轴与标签（X轴0-14000、城市标签位置）
plt.xlim(0, 14000)
plt.xticks(range(0, 14001, 2000))
plt.xlabel('人均消费支出（元）', fontsize=12)
plt.ylabel('城市', fontsize=12)
plt.yticks(y_pos, cities)
plt.title('4个城市两季度人均消费支出对比', fontsize=14, fontweight='bold', pad=20)
plt.legend(loc='lower right')

# 调整布局
plt.tight_layout()
plt.savefig('图表2_城市季度消费横向条形图.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 还原图表数据（匹配Y轴60-130范围、异常值位置）
np.random.seed(42)  # 固定随机种子，确保异常值位置一致
data_a = np.random.normal(loc=100, scale=5, size=50)  # 工艺A：集中在90-110
data_b = np.random.normal(loc=105, scale=3, size=50)  # 工艺B：集中在95-115（无异常）
data_c = np.concatenate([  # 工艺C：含130、60两个异常值（匹配图表）
    np.random.normal(loc=95, scale=8, size=48),
    np.array([130, 60])
])
all_data = [data_a, data_b, data_c]
labels = ['A', 'B', 'C']  # X轴工艺标签

# 创建画布
plt.figure(figsize=(8, 6), dpi=100)

# 绘制箱形图（匹配图表样式：显示异常值、箱形颜色）
box_plot = plt.boxplot(
    all_data, tick_labels=labels, patch_artist=True,
    showfliers=True,  # 显示异常值（60、130）
    medianprops={'color': 'black', 'linewidth': 1.5},  # 中位数线（黑色）
    flierprops={'marker': 'o', 'color': 'red', 'markersize': 6}  # 异常值（红色圆点）
)

# 箱形填充颜色（匹配图表浅色系）
colors = ['#E2EFDA', '#D9E2F3', '#FBE5D6']
for patch, color in zip(box_plot['boxes'], colors):
    patch.set_facecolor(color)

# 还原坐标轴与标签（Y轴60-130范围、“质量（克）”标签）
plt.ylim(60, 130)
plt.yticks(range(60, 131, 10))
plt.xlabel('生产工艺', fontsize=12)
plt.ylabel('零件质量（克）', fontsize=12)
plt.title('三组工艺生产零件质量分布', fontsize=14, fontweight='bold', pad=20)

# 调整布局
plt.tight_layout()
plt.savefig('图表3_零件质量箱形图.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 还原图表数据（匹配X轴400-1400、Y轴0-700、气泡颜色/大小）
cities = ['A', 'B', 'C', 'D', 'E', 'F']
population = [300, 1500, 800, 500, 1200, 900]  # X轴人口（匹配400-1400范围）
gdp = [150, 700, 350, 200, 600, 450]          # Y轴GDP（匹配0-700范围）
income = [6.0, 9.0, 7.5, 6.5, 8.5, 7.0]       # 气泡大小/颜色（匹配6.0-9.0标注）

# 气泡大小缩放（匹配图表气泡尺寸）
bubble_size = np.array(income) * 300  # 缩放因子300，避免气泡过大/过小

# 创建画布
plt.figure(figsize=(10, 6), dpi=100)

# 绘制气泡图（匹配图表样式：颜色映射、黑色边框）
scatter = plt.scatter(
    x=population, y=gdp, s=bubble_size, c=income,
    cmap='viridis', alpha=0.7, edgecolors='black', linewidth=1
)

# 还原城市标签（匹配图表A-F位置）
for i, city in enumerate(cities):
    plt.text(
        x=population[i] + 30, y=gdp[i], s=city,
        ha='left', va='center', fontsize=11, fontweight='bold'
    )

# 还原坐标轴与标签（X轴400-1400、Y轴0-700、颜色条标注）
plt.xlim(400, 1400)
plt.ylim(0, 700)
plt.xticks(range(400, 1401, 200))
plt.yticks(range(0, 701, 100))
plt.xlabel('人口（万人）', fontsize=12)
plt.ylabel('GDP（亿元）', fontsize=12)
plt.title('不同城市人口、GDP与人均收入关系', fontsize=14, fontweight='bold', pad=20)

# 添加颜色条（匹配图表右侧“人均收入（万元）”标注）
cbar = plt.colorbar(scatter)
cbar.set_label('人均收入（万元）', fontsize=11)
cbar.set_ticks([6.0, 7.0, 8.0, 9.0])  # 匹配图表颜色标注

# 调整布局
plt.tight_layout()
plt.savefig('图表7_城市人口GDP气泡图.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 还原图表数据（匹配X轴1-12小时、Y轴0-60强度、误差棒）
time = [1, 2, 4, 8, 12]  # X轴处理时间（非连续，匹配图表）
mean_intensity = [15, 22, 35, 48, 55]  # 平均强度（0-60，递增趋势）
std_dev = [1.5, 2.0, 3.5, 4.0, 5.5]    # 误差棒长度（随时间增大，匹配图表）

# 创建画布
plt.figure(figsize=(10, 6), dpi=100)

# 绘制误差棒图（匹配图表样式：蓝色实线、圆形标记、红色误差棒）
plt.errorbar(
    x=time, y=mean_intensity, yerr=std_dev,
    fmt='o-',  # 圆形标记+实线连接
    color='#4472C4', linewidth=2, markersize=6,
    ecolor='red', elinewidth=1.5, capsize=4, capthick=1.5  # 红色误差棒
)

# 还原坐标轴与标签（X轴1-12小时、Y轴0-60强度、“平均反应强度”标签）
plt.xlim(0, 13)
plt.ylim(0, 60)
plt.xticks([1, 2, 4, 8, 12])  # 仅显示数据点对应的X刻度
plt.yticks(range(0, 61, 10))
plt.xlabel('处理时间（小时）', fontsize=12)
plt.ylabel('平均反应强度', fontsize=12)
plt.title('不同处理时间下实验平均反应强度', fontsize=14, fontweight='bold', pad=20)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 调整布局
plt.tight_layout()
plt.savefig('图表9_实验反应强度误差棒图.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 还原图表数据（匹配5个指标、0-10分、产品A/B得分差异）
categories = ['性能', '设计', '续航', '价格', '服务']  # 5个指标（匹配图表顺序）
product_a = [9.0, 7.5, 8.0, 6.5, 9.0]  # 产品A得分（性能高、价格低，匹配图表）
product_b = [7.0, 8.5, 7.5, 8.0, 9.0]  # 产品B得分（设计高、价格高，匹配图表）
n = len(categories)

# 雷达图角度计算（闭合图形）
angles = np.linspace(0, 2 * np.pi, n, endpoint=False).tolist()
product_a = product_a + [product_a[0]]
product_b = product_b + [product_b[0]]
angles = angles + [angles[0]]
categories = categories + [categories[0]]

# 创建画布（极坐标）
plt.figure(figsize=(8, 8), dpi=100)
ax = plt.subplot(111, polar=True)

# 绘制雷达图（匹配图表样式：实线/虚线、填充色）
ax.plot(
    angles, product_a, marker='o', linestyle='-',
    color='#4472C4', linewidth=2, markersize=6, label='产品A'
)
ax.fill(angles, product_a, color='#4472C4', alpha=0.2)  # 产品A填充色

ax.plot(
    angles, product_b, marker='s', linestyle='--',
    color='#ED7D31', linewidth=2, markersize=6, label='产品B'
)
ax.fill(angles, product_b, color='#ED7D31', alpha=0.2)  # 产品B填充色

# 还原坐标轴与标签（0-10分、指标标签位置）
ax.set_ylim(0, 10)
ax.set_yticks(range(0, 11, 2))
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories[:-1], fontsize=11)
ax.set_title('两款产品关键性能指标对比', fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))

# 调整布局
plt.tight_layout()
plt.savefig('图表10_产品性能雷达图.png', dpi=300)
plt.show()
```
