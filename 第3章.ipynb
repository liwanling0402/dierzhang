# 3.2　设置坐标轴的标签、刻度范围和刻度标签

## 1.单组柱形图

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 加载数据
categories = ['电子产品', '服装鞋帽', '家居百货', '美妆个护', '食品生鲜']
sales = [550, 420, 280, 390, 610]  # 对应销售额（万元）
bar_color = '#4a90e2'  # 接近原图的蓝绿色

# 2. 创建画布
plt.figure(figsize=(8, 6))

# 3. 绘制柱状图
bars = plt.bar(categories, sales, color=bar_color, width=0.6)

# 4. 添加数据标签（在柱子上方显示销售额）
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 10,
             f'{height}', ha='center', va='bottom', fontsize=10)

# 5. 设置图表标题与标签
plt.title('1. 各品类月销售额统计（单组柱形图）', fontsize=14, pad=20)
plt.xlabel('商品品类', fontsize=12)
plt.ylabel('销售额（万元）', fontsize=12)

# 6. 调整样式
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(sales) + 100)  # 调整y轴范围，避免数据标签超出
plt.tight_layout()

# 7. 显示图表
plt.show()
```

## 2.多组柱形图

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 数据（匹配图中y轴城市顺序）
cities = ['深圳', '广州', '上海', '北京']
q1_spend = [10500, 9800, 11500, 12000]  # Q1支出
q2_spend = [11900, 10900, 12800, 13500]  # Q2支出
bar_width = 0.35  # 条形宽度
y_pos = np.arange(len(cities))  # y轴坐标


# 2. 创建画布并绘图
plt.figure(figsize=(10, 5))  # 匹配示例的图尺寸

# 绘制Q1、Q2的水平条形（颜色与示例一致：红色系+蓝绿色系）
plt.barh(y_pos - bar_width/2, q1_spend, bar_width, 
         color='#e74c3c', label='Q1支出')
plt.barh(y_pos + bar_width/2, q2_spend, bar_width, 
         color='#3498db', label='Q2支出')


# 3. 图表样式配置（完全对齐示例）
plt.yticks(y_pos, cities)  # y轴显示城市名
plt.xlabel('人均消费支出（元）')
plt.title('2. 不同城市季度人均消费支出对比')
plt.legend(loc='lower right')  # 图例位置与示例一致

# 调整x轴刻度范围（避免条形过满）
plt.xlim(0, 15000)
plt.tight_layout()  # 自动调整布局


# 4. 显示图表
plt.show()

```

## 3.箱型图

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 生成各工艺的质量数据
np.random.seed(42)  # 固定随机种子，确保结果可复现

# 工艺A：均值100，标准差5，50个样本
data_A = np.random.normal(loc=100, scale=5, size=50)
# 工艺B：均值105，标准差3，50个样本
data_B = np.random.normal(loc=105, scale=3, size=50)
# 工艺C：均值95，标准差8，48个样本 + 2个异常值
data_C = np.concatenate([np.random.normal(loc=95, scale=8, size=48), [130, 60]])


# 2. 创建画布并绘制箱形图
plt.figure(figsize=(8, 6))
# 绘制箱形图（匹配图中样式：蓝色箱体、显示异常值）
boxplot = plt.boxplot([data_A, data_B, data_C], 
                      labels=['工艺A', '工艺B', '工艺C'],
                      patch_artist=True,  # 允许填充箱体颜色
                      flierprops={'marker': 'o', 'color': 'black'})  # 异常值样式


# 3. 设置箱体颜色（匹配图中蓝色）
for patch in boxplot['boxes']:
    patch.set_facecolor('#3498db')


# 4. 图表样式配置
plt.ylabel('质量（克）')
plt.title('3. 不同工艺下零件质量分布（箱形图）')
plt.ylim(50, 140)  # 匹配图中y轴范围
plt.grid(axis='y', linestyle='--', alpha=0.7)  # 添加y轴网格线


# 5. 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 年份数据
years = [2019, 2020, 2021, 2022, 2023, 2024, 2025]
# 营收数据（亿元）
revenues = [2.5, 3.1, 4.0, 5.2, 6.5, 7.8, 8.5]
# 利润数据（亿元）
profits = [0.5, 0.7, 0.9, 1.3, 1.8, 2.2, 2.5]

# 创建画布
plt.figure(figsize=(8, 6))
# 绘制营收折线
plt.plot(years, revenues, marker='o', label='营收（亿元）', color='blue')
# 绘制利润折线
plt.plot(years, profits, marker='o', label='利润（亿元）', color='orange', linestyle='--')

# 添加标题和坐标轴标签
plt.title('公司近7年营收和利润变化（折线图）')
plt.xlabel('年份')
plt.ylabel('金额（亿元）')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
